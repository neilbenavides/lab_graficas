```python
import numpy as np

rng = np.random.default_rng(42)

# Datos de calificaciones de los estudiantes
matematicas = rng.integers(50, 100, 20)
ciencias = rng.integers(40, 95, 20)
literatura = rng.integers(60, 100, 20)

# Datos de errores para el gr√°fico de barras de error
errores_matematicas = rng.uniform(2, 5, 2)
errores_matematicas = [min(errores_matematicas), max(errores_matematicas)]

errores_ciencias = rng.uniform(1, 4, 2)
errores_ciencias = [min(errores_ciencias), max(errores_ciencias)]

errores_literatura = rng.uniform(3, 6, 2)
errores_literatura = [min(errores_literatura), max(errores_literatura)]
```

```python
import matplotlib.pyplot as plt
# Crear el gr√°fico de dispersi√≥n
plt.scatter(matematicas, ciencias)

# Personalizar el gr√°fico
plt.title('Relaci√≥n entre las calificaciones de Matem√°ticas y Ciencias')
plt.xlabel('Calificaciones de Matem√°ticas')
plt.ylabel('Calificaciones de Ciencias')

# Mostrar el gr√°fico
plt.show()
```

```python
# Calcular las calificaciones promedio
promedios = [np.mean(matematicas), np.mean(ciencias), np.mean(literatura)]

# Calcular los errores
errores = [errores_matematicas, errores_ciencias, errores_literatura]
errores_promedio = [np.mean(err) for err in errores]


# Crear el gr√°fico de barras de error
plt.errorbar(['Matem√°ticas', 'Ciencias', 'Literatura'], promedios, yerr=errores_promedio, fmt='o', capsize=5)

# Personalizar el gr√°fico
plt.title('Calificaciones promedio y errores en Matem√°ticas, Ciencias y Literatura')
plt.xlabel('Materias')
plt.ylabel('Calificaciones promedio')

# Mostrar el gr√°fico
plt.show()
```

```python
plt.hist(matematicas, bins=10, edgecolor='black')

# Personalizar el histograma
plt.title('Distribuci√≥n de las calificaciones de Matem√°ticas')
plt.xlabel('Calificaciones')
plt.ylabel('Cantidad de estudiantes')

# Mostrar el histograma
plt.show()
```
